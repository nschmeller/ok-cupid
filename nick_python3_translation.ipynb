```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
profile = pd.read_csv('profiles.csv')
```

```python
profile
```

```python
# remove -1 (prefer not to say) response from income column
profile_trimmed = profile[profile['income'] != -1]
```

```python
# one-hot encoding for categorical data
categorical_columns = ['body_type', 'diet', 'drinks', 'drugs', 'education', 'ethnicity', 'job', 
                      'offspring', 'orientation', 'pets', 'religion', 'sex', 'sign',
                      'smokes', 'speaks', 'status']
profile_trimmed_encoded = pd.get_dummies(profile_trimmed, columns=categorical_columns)
```

```python
np.unique(profile_trimmed_encoded.dtypes)
```

```python
num_cols = [c for c in profile_trimmed_encoded.columns.values if 
           (profile_trimmed_encoded[c].dtype == 'uint8') or
           (profile_trimmed_encoded[c].dtype == 'int64') or
           (profile_trimmed_encoded[c].dtype == 'float64')]
profile_trimmed_encoded = profile_trimmed_encoded[num_cols]
```

```python
%%time

from sklearn.decomposition import LatentDirichletAllocation

lda = LatentDirichletAllocation(n_components=5, learning_method='online')
lda.fit(profile_trimmed_encoded)
```
